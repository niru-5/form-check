```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import os
```

```python
data_folder = '/hdd/side_projects/imu_project/form-check/data_acquisition/data'

folder_name = 'completely_still_6'

# list all csv files in the data folder
csv_files = sorted([f for f in os.listdir(os.path.join(data_folder, folder_name)) if f.endswith('.csv')])

# load the first csv file
print (csv_files)
```

```python
folder_names = ['completely_still_4', 'completely_still_5', 'completely_still_6', 'completely_still_7', 'completely_still_8', 'completely_still_9']
run_names = ['still', 'still', 'slow_speed', 'slow_speed', 'med_speed', 'high_speed']


# graphs for each run, 2*3 subplots for acc and gyro 


# accumulate all the acc data
acc_data = []
for i in range(len(folder_names)):
    folder_name = folder_names[i]
    run_name = run_names[i]
    csv_files = sorted([f for f in os.listdir(os.path.join(data_folder, folder_name)) if f.endswith('.csv') and f.startswith('acc')])
    acc_df = pd.read_csv(os.path.join(os.path.join(data_folder, folder_name), csv_files[0]))
    acc_data.append(acc_df)

# Create a 3x2 subplot figure
fig, axs = plt.subplots(3, 2, figsize=(15, 20))

# Flatten axs array for easier iteration
axs_flat = axs.flatten()

# Set major and minor grid lines
for ax in axs_flat:
    ax.set_ylim(-1.2, 1.2)
    ax.yaxis.set_major_locator(plt.MultipleLocator(0.2))
    ax.yaxis.set_minor_locator(plt.MultipleLocator(0.1))

# Plot each run's acceleration data
for i in range(len(acc_data)):
    df = acc_data[i]
    ax = axs_flat[i]
    
    ax.plot(df['epoch'], df['x'], label='X')
    ax.plot(df['epoch'], df['y'], label='Y') 
    ax.plot(df['epoch'], df['z'], label='Z')
    
    ax.set_title(f'Run: {run_names[i]}')
    ax.legend()
    ax.grid(True)

# Remove empty subplot in bottom right since we only have 6 plots
if len(acc_data) < 6:
    fig.delaxes(axs_flat[5])

plt.tight_layout()
plt.show()


```
